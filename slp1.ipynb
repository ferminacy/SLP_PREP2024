```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
# Step 1: Combine q29, q30, q31 into FrequencyScore
df_cleaned['FrequencyScore'] = df_cleaned[['q29', 'q30', 'q31']].sum(axis=1)

# Step 2: Categorize Frequency Levels
def map_frequency(score):
    if score <= 6:  # Example threshold for "Low Frequency" (adjust as needed)
        return "Low"
    elif score <= 9:  # Example threshold for "Medium Frequency"
        return "Medium"
    else:
        return "High"  # Example threshold for "High Frequency"

df_cleaned['FrequencyLevel'] = df_cleaned['FrequencyScore'].apply(map_frequency)

# Step 3: Map ExperienceGroup (Years of Experience)
experience_mapping = {
    1: "Low Experience",  # Corresponds to a, b, c
    2: "Moderate Experience",  # Corresponds to d, e
    3: "High Experience"  # Corresponds to f, g
}
df_cleaned['ExperienceGroup'] = df_cleaned['q11'].map(experience_mapping)

# Step 4: Create a Contingency Table
contingency_table = pd.crosstab(df_cleaned['ExperienceGroup'], df_cleaned['FrequencyLevel'])
print("Contingency Table:")
print(contingency_table)

# Step 5: Perform Chi-Square Test
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Step 6: Output Results
print("\nChi-Square Test Results:")
print(f"Chi-Square Statistic: {chi2}")
print(f"P-Value: {p}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)
```

The results of the chi-square test (p-value = 0.073) show that there is no statistically significant association between years of experience and the frequency of interpreter use. However, the p-value's proximity to the 0.05 threshold suggests that additional factors may influence interpreter use that were not included in this analysis.

```python
import pandas as pd
from scipy.stats import f_oneway

# Step 1: Calculate Preparedness Score by summing Q32, Q34, and Q36
df_cleaned['PreparednessScore'] = df_cleaned[['q32', 'q34', 'q36']].sum(axis=1)

# Step 2: Group by Frequency of Workshops (Q54)
groups = [df_cleaned[df_cleaned['q54'] == value]['PreparednessScore'] for value in df_cleaned['q54'].unique()]

# Step 3: Perform One-Way ANOVA
f_stat, p_value = f_oneway(*groups)

# Step 4: Output Results
print("One-Way ANOVA Results:")
print(f"F-Statistic: {f_stat}")
print(f"P-Value: {p_value}")

# Optional: Visualize the results
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='q54', y='PreparednessScore', data=df_cleaned)
plt.xlabel('Frequency of Workshops (Q54)')
plt.ylabel('Preparedness Score')
plt.title('Preparedness Score by Frequency of Workshops')
plt.show()
```

The one-way ANOVA results (F = 10.31, p < 0.001) show a significant link between how often workshops were offered during graduate programs and how prepared participants feel to work with interpreters. Those who attended workshops more frequently (e.g., "Often" or "Always") reported higher preparedness scores compared to those who attended fewer workshops (e.g., "Never" or "Rarely").

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Step 1: Compute the Preparedness Score
df_cleaned['PreparednessScore'] = df_cleaned[['q32', 'q34', 'q36']].sum(axis=1)

# Step 2: Perform Two-Way ANOVA
model = ols('PreparednessScore ~ C(q45) * C(q54)', data=df_cleaned).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Display ANOVA Results
print("Two-Way ANOVA Results:")
print(anova_table)

# Optional: Visualize Interaction Effect
sns.boxplot(
    x='q54', y='PreparednessScore', hue='q45', data=df_cleaned, palette='Set3'
)
plt.title("Preparedness Score by Workshop Frequency and Quality")
plt.xlabel("Frequency of Workshops (Q54)")
plt.ylabel("Preparedness Score")
plt.legend(title="Quality of Workshops (Q45)", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

The frequency of workshops (Q54) significantly impacts participants' preparedness scores, as indicated by a significant p-value (< 0.05). More frequent workshops contribute to higher preparedness levels.

The quality of workshops (Q45) does not independently influence preparedness, with a p-value of 1.000.

The interaction between frequency and quality (Q54 * Q45) not statistically significant, but the p-value of 0.064 suggests a trend worth investigating with a larger dataset or refined measures.
